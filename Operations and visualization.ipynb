Getting stock prices between 1 January 2015 and 30 December 2016

```python
nifty = get_history(symbol="NIFTYIT",start=date(2015,1,1),end=date(2016,12,30),index=True)
```

```python
tcs = get_history(symbol="TCS",start=date(2015,1,1),end=date(2016,12,30))
```

```python
tcs.shape
```

```python
infy = get_history(symbol="INFY",start=date(2015,1,1),end=date(2016,12,30))
```

```python
infy.shape
```

```python
tcs_dates = tcs.index.values.tolist()
infy_dates = infy.index.values.tolist()
nifty_dates = nifty.index.values.tolist()
```

```python
tcs_week = []
for i in range(len(tcs_dates)):
    tcs_week.append(int(tcs_dates[i].strftime("%V")))
    
infy_week = []
for i in range(len(tcs_dates)):
    infy_week.append(int(infy_dates[i].strftime("%V")))
    
nifty_week = []
for i in range(len(tcs_dates)):
    nifty_week.append(int(nifty_dates[i].strftime("%V")))
```

```python
tcs['week'] = tcs_week
infy['week'] = infy_week
nifty['week'] = nifty_week
```

```python
week_bias = {'2015':0,
            '2016':52
            }

# FOR TCS
week_number  = []
years = []
for i in range(len(tcs_dates)):
    year = tcs_dates[i].strftime("%Y")
    years.append(year)
    week_number.append(tcs_week[i] + week_bias[year])
tcs['week_number'] = week_number
tcs['year'] = years

# FOR INFY
week_number  = []
years = []
for i in range(len(infy_dates)):
    year = infy_dates[i].strftime("%Y")
    years.append(year)
    week_number.append(infy_week[i] + week_bias[year])
infy['week_number'] = week_number
infy['year'] = years

# FOR NIFTY
week_number  = []
years = []
for i in range(len(nifty_dates)):
    year = nifty_dates[i].strftime("%Y")
    years.append(year)
    week_number.append(nifty_week[i] + week_bias[year])
nifty['week_number'] = week_number
nifty['year'] = years
```

```python
# Function for calculating moving averages
def movingaverage(data, window_size):
    week_data = data.groupby("week_number")["Close"].mean().values
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(week_data, window, 'valid')
```

```python
(movingaverage(tcs, 4))
```

```python
(movingaverage(infy, 4))
```

```python
(movingaverage(nifty, 4))
```

Price Shocks and directions

```python
##  TCS 
## Price Shock
tcs["price_shock"] = (((abs(tcs["Close"] - tcs["Prev Close"])/tcs["Prev Close"])*100) > 2).astype(int)
## Shock Direction 
tcs["price_shock_direction"] = (tcs["Close"] - tcs["Prev Close"] > 0).astype(int)

##  INFY 
## Price Shock
infy["price_shock"] = (((abs(infy["Close"] - infy["Prev Close"])/infy["Prev Close"])*100) > 2).astype(int)
## Shock Direction 
infy["price_shock_direction"] = (infy["Close"] - infy["Prev Close"] > 0).astype(int)

```

Volume Shock and Directions

```python
## TCS
## Volume Shock
tcs["volume_shock"] = ((abs((tcs["Volume"].shift(1) - tcs["Volume"])/tcs["Volume"].shift(1))*100)  > 10).astype(int)
## Shock Direction
tcs['volume_shock_direction'] = (tcs["Volume"].shift(1) - tcs["Volume"] > 0).astype(int)

## INFY
## Volume Shock
infy["volume_shock"] = ((abs((infy["Volume"].shift(1) - infy["Volume"])/infy["Volume"].shift(1))*100)  > 10).astype(int)
## Shock Direction
infy['volume_shock_direction'] = (infy["Volume"].shift(1) - infy["Volume"] > 0).astype(int)

```

Price shock without Volume Shock

```python
## TCS

tcs["not_volume_shock"]  = (~(tcs["volume_shock"].astype(bool))).astype(int)
tcs["price_shock_without_volume_shock"] = tcs["not_volume_shock"] & tcs["price_shock"]

## INFY

infy["not_volume_shock"]  = (~(infy["volume_shock"].astype(bool))).astype(int)
infy["price_shock_without_volume_shock"] = infy["not_volume_shock"] & infy["price_shock"]
```

```python
## Time-series for close price

## TCS
from bokeh.plotting import figure, output_notebook, show
tcs_2015 = tcs[tcs["year"] == "2015"]
tcs_with_shock = tcs_2015[tcs_2015["volume_shock"] == 1]
y1 = tcs_with_shock["Close"].values
x1 = tcs_with_shock.index.values
output_notebook()
x = tcs_2015.index.values
y = tcs_2015["Close"].values
p = figure(title="closing_price", x_axis_label='x', y_axis_label='y')
p.line(x, y, legend="Close Price", line_color="blue")
p.line(x, y1,legend="Close price for volume shocks",line_color="red", line_dash="4 4")
show(p)
```

```python
## TCS
from statsmodels.tsa.stattools import acf, pacf
import bokeh
# from bokeh.charts import Bar
from bokeh.io import output_notebook, show
output_notebook()
lags = 50
x = list(range(lags))
p = figure(plot_height=350, title="Partial Autocorrelation")
partial_autocorr = pacf(tcs[tcs["year"] == "2015"]["Close"], nlags=lags)
p.vbar(x=x, top=partial_autocorr, width=0.9)
show(p)
```
